```python
import pandas as pd
```

```python
import os

path = './res_gossiping'
file_list = os.listdir(path)
file_list
```

```python
os.listdir('./')
```

```python
columns = ['推', '噓', '分數', '作者', '標題', '時間']
```

```python
data = []
for file_name in file_list:
    try:
        with open(path + '/' + file_name, 'r', encoding='utf-8') as f:
            tmp_data = f.read().split('---split---')[1].split('\n')
    #         print(tmp_data)
        tmp_data = tmp_data[1:-1]
        data.append(tmp_data)
#         print(tmp_data)
    except:
        pass
data
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def column_filter(s):
    output = s.split(': ')[-1]
    return output
```
