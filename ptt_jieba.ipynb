```python
import jieba
import os
```

```python
path = './res_gossiping'
file_list = os.listdir(path)
file_list
```

```python
jieba.load_userdict('./mydict.txt')
```

```python
s = ''
for file_name in file_list:
    with open(path + '/' + file_name, 'r') as f:
        s += f.read().split('---split---')[0]
s
```

```python
s_list = jieba.cut(s)
```

```python
s_dict = {}
for w in s_list:
    if len(w.strip()) < 2:
        continue
    if w in s_dict:
        s_dict[w] += 1
    else:
        s_dict[w] = 1
s_dict
```

```python
tmp_list = []
for i in s_dict:
    tmp_list.append((i, s_dict[i]))
tmp_list
```

```python
tmp_list.sort(key=lambda x: x[1], reverse=True)
tmp_list
```
